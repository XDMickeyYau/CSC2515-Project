```python
image = np.load("simCLRfeature/CIFARDataTest.npz")['arr_0'].reshape(-1,32,32,3)
image.shape
```

```python
feature = np.load("simCLRfeature/CIFARDataTestFeature.npz")['arr_0']
feature.shape
```

```python
label = np.array([i for i in range(4) for j in range(1000)])
label[::1000]
```

```python
n_cluster = 10
kmeans_model, data_index_cluster_labels = generate_kmeans_model(feature, n_cluster)
cluster_index_dict = clusters_to_index(data_index_cluster_labels)
```

```python
# Vidualize image clusters
for cluster in sorted(list(cluster_index_dict.keys())):
  print(cluster)
  image_index = cluster_index_dict[cluster]
  images = image[image_index,:,:,:]
  visualize_data(3,3, 5, images)
```

```python
from sklearn import metrics
metrics.normalized_mutual_info_score(label, kmeans_model.labels_)
```
